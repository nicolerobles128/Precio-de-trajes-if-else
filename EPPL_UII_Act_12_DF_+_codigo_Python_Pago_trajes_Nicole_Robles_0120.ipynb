```python
# Calculando el descuento y precio final usando if else
# Datos de entrada CT (Costo del Traje)
# Si es verdadero (CT > 2500) aplica 15%
# De lo contrario aplica 8%

CT = float(input("Ingrese el costo del traje: "))

if CT > 2500:
    DE = CT * 0.15
else:
    DE = CT * 0.08

PF = CT - DE

print("El descuento es:", DE)
print("El precio final es:", PF)

print("Fin del programa")
print("Programa realizado por Nicole Robles NL = 46")
```
